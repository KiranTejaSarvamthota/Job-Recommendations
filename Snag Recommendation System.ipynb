```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

rating_headers = ['user_id', 'job_id', 'rating', 's.no']
ratings = pd.read_table('input files/snagratings.csv', sep=',', header=None, names=rating_headers)
```

```python
with open('input files/snagjobs.csv', mode='r') as infile:
    reader = csv.reader(infile)
    jobsdict = {rows[0]:rows[1] for rows in reader}
```

```python
ratings.head()
```

```python
rp = ratings.pivot_table(columns=['job_id'],index=['user_id'],values='rating')
rp.head()
```

```python
rp = rp.fillna(0); # Replace NaN
rp.head()
```

```python
jobs_sorted = list(rp)
```

```python
users_sorted= rp.index.values
```

```python
Q = rp.values
Q
```

```python
W = Q>0.5
W[W == True] = 1
W[W == False] = 0
# To be consistent with our Q matrix
W = W.astype(np.float64, copy=False)
```

```python
W
```

```python
lambda_ = 0.1
n_factors = 10
m, n = Q.shape
n_iterations = 50
```

```python
X = 5 * np.random.rand(m, n_factors) 
Y = 5 * np.random.rand(n_factors, n)
```

```python
def get_error(Q, X, Y, W):
    return np.sum((W * (Q - np.dot(X, Y)))**2)
```

```python
errors = []
for ii in range(n_iterations):
    X = np.linalg.solve(np.dot(Y, Y.T) + lambda_ * np.eye(n_factors), 
                        np.dot(Y, Q.T)).T
    Y = np.linalg.solve(np.dot(X.T, X) + lambda_ * np.eye(n_factors),
                        np.dot(X.T, Q))
    if ii % 100 == 0:
        print('{}th iteration is completed'.format(ii))
    errors.append(get_error(Q, X, Y, W))
Q_hat = np.dot(X, Y)
print('Error of rated movies: {}'.format(get_error(Q, X, Y, W)))
```

```python
plt.plot(errors);
plt.ylim([0, 40000]);
```

```python
def print_recommendations(W=W, Q=Q, Q_hat=Q_hat):

    Q_hat -= np.min(Q_hat)

    output=open("output file/Snag Recommendations output.txt",'w')
    Q_hat *= float(5) / np.max(Q_hat)

    jobs_ids = np.argsort(Q_hat - 5 * W, axis=1)

    for jj in range (m):
        job_id = jobs_ids[jj]

        number_of_recs = 5
        string_recommend="\n"+"-------------------Recommendations for user"+"  "+users_sorted[jj]+"--------------------------"+"\n"
        for j_id in job_id[::-1]:
            
            #string_recommend=string_recommend+"Job title"+" :  "+str(job_titles[j_id])+","+"Predicted Rating: "+"kk"+"\n"
            
            string_recommend = string_recommend+ '\n Recommended  Job Title {} and Job Id {}- with predicted rating: {}'.format(
                jobsdict[jobs_sorted[j_id]], jobs_sorted[j_id], Q_hat[jj, j_id])
            string_recommend= string_recommend +'\n'
            number_of_recs -= 1
            if number_of_recs == 0:
                break
                
        final_string=string_recommend.encode('utf-8', 'ignore')
        output.write(final_string.decode(output.encoding))
    print('Job Recommendations are saved to the output file')
```

```python
weighted_errors = []
for ii in range(n_iterations):
    for u, Wu in enumerate(W):
        X[u] = np.linalg.solve(np.dot(Y, np.dot(np.diag(Wu), Y.T)) + lambda_ * np.eye(n_factors),
                               np.dot(Y, np.dot(np.diag(Wu), Q[u].T))).T
    for i, Wi in enumerate(W.T):
        Y[:,i] = np.linalg.solve(np.dot(X.T, np.dot(np.diag(Wi), X)) + lambda_ * np.eye(n_factors),
                                 np.dot(X.T, np.dot(np.diag(Wi), Q[:, i])))
    weighted_errors.append(get_error(Q, X, Y, W))
    print('{}th iteration is completed'.format(ii))
weighted_Q_hat = np.dot(X,Y)
print('Error of rated jobs: {}'.format(get_error(Q, X, Y, W)))
```

```python
plt.plot(weighted_errors);
plt.xlabel('Iteration Number');
plt.ylabel('Mean Squared Error');
```

```python
print(weighted_Q_hat)
```

```python
print_recommendations(Q_hat=weighted_Q_hat)
```
